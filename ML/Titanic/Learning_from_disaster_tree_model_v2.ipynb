## Importing and loading data

```python
import pandas as pd

test_path = r'C:\Users\tomek\Documents\titanic_test.csv'
train_path = r'C:\Users\tomek\Documents\titanic_train.csv'

train_data = pd.read_csv(train_path)
test_data = pd.read_csv(test_path)
```

```python
train_data.head(n = 20)
```

### Calculating mean age values for each sex and replacing null values with them

```python
mean_man_age = train_data.loc[(train_data.Sex == 'male') & (train_data.Age.notna())]

mean_man_age.Age.mean().round()
```

```python
mean_female_age = train_data.loc[(train_data.Sex == 'female') & (train_data.Age.notna())]

mean_female_age.Age.mean().round()
```

```python
train_data.loc[(train_data.Sex == 'male') & (train_data.Age.isna()), 'Age'] = 31
train_data.loc[(train_data.Sex == 'female') & (train_data.Age.isna()), 'Age'] = 28

train_data.head(n = 20)
```

### Replacing sex labels (object type) in order to use it as one of model's feature

```python
train_data = pd.get_dummies(train_data, columns=['Sex'], prefix=['Sex'])
```

```python
train_data.head()
```

```python
from sklearn.tree import DecisionTreeRegressor

train_model = DecisionTreeRegressor(random_state=0)

y = train_data.Survived
features2 = ['Age', 'Pclass', 'Sex_female', 'Sex_male']
features1 = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_female', 'Sex_male']

X = train_data[features1]

train_model.fit(X, y)
```

```python
pred_y = train_model.predict(X).round()
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y, pred_y)

mae
```

```python
test_data
```

```python
test_data = pd.get_dummies(test_data, columns=['Sex'], prefix = ['Sex'])
```

```python
test_X = test_data[features1]

pred_test_y = train_model.predict(test_X).round()
```

```python
output = pd.DataFrame({'PassengerID': test_data.PassengerId, 'Survived': pred_test_y.astype(int)})
print(output)
output.to_csv('titanic_model2.csv', index = False)
```
