```python
import pandas as pd

test_path = r'C:\Users\tomek\Documents\titanic_test.csv'
train_path = r'C:\Users\tomek\Documents\titanic_train.csv'

test_data = pd.read_csv(test_path)
train_data = pd.read_csv(train_path)
```

```python
train_data.head(n = 20)
```

```python
male_ageless_psgrs = train_data.loc[(train_data.Sex == 'male') & (train_data.Age.notna())]

mean_m_age = male_ageless_psgrs.Age.mean().round()
```

```python
female_ageless_psgrs = train_data.loc[(train_data.Sex == 'female') & (train_data.Age.notna())]

mean_f_age = female_ageless_psgrs.Age.mean().round()
```

```python
train_data.loc[(train_data.Sex == 'male') & (train_data.Age.isna()), 'Age'] = mean_m_age
train_data.loc[(train_data.Sex == 'female') & (train_data.Age.isna()), 'Age'] = mean_f_age
```

```python
def chad(row):
    if row.Age <= 18:
        return('child')
    elif row.Age > 18:
        return('adult')
    
train_data['child/adult'] = train_data.apply(chad, axis = 'columns')
```

```python
train_data = pd.get_dummies(train_data, columns = ['Sex'], prefix = ['Sex'])
```

```python
train_data = pd.get_dummies(train_data, columns = ['child/adult'], prefix = ['chad'])
```

```python
train_data.head(n = 20)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error

tree_model = DecisionTreeRegressor(random_state = 1)

y = train_data.Survived
features_set = ['Pclass', 'Age', 'SibSp', 'Fare', 'Sex_female', 'Sex_male', 'chad_adult', 'chad_child']
X = train_data[features_set]

tree_model.fit(X, y)
```

```python
pred_y = tree_model.predict(X)
```

```python
mae = mean_absolute_error(y, pred_y)
mae
```

```python
test_data['child/adult'] = test_data.apply(chad, axis = 'columns')
```

```python
test_data.loc[test_data.Age <= 18, 'child/adult'] = 'child'
test_data.loc[test_data.Age > 18, 'child/adult'] = 'adult'
```

```python
test_data = pd.get_dummies(test_data, columns = ['Sex'], prefix = ['Sex'])
```

```python
test_data = pd.get_dummies(test_data, columns = ['child/adult'], prefix = ['chad'])
```

```python
test_data
```

```python
test_X = test_data[features_set]
pred_test_y = tree_model.predict(test_X).round()
```

```python
output = pd.DataFrame({'PassengerId' : test_data.PassengerId, 'Survived' : pred_test_y.astype(int)})
```

```python
output.to_csv('titanic_model3.csv', index = False)
```
